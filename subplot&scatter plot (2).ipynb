```python
import numpy as np
import matplotlib.pyplot as plt
x1=[1,5,7,8,9,5]
y1=[2,5,6,8,9,7]
plt.subplot(1,2,1)
plt.plot(x1,y1)
plt.title("Graph1")
#subplot
x2=[9,6,5,4,7,2]
y2=[8,5,6,9,7,1]
plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.title("Graph2")
plt.suptitle("Graph")
```

```python
#scatterplot

x1=[1,4,3,5,6]
y1=[2,4,8,9,5]
plt.scatter(x1,y1)

x2=[5,6,9,8,7]
y2=[2,5,6,9,8]
plt.scatter(x2,y2)
```

```python
x1=[2,4,5,3,9]
y1=[8,5,6,9,8]
plt.bar(x1,y1)
```

```python
x1=[2,4,5,3,9]
y1=[8,5,6,9,8]
plt.subplot(1,2,1)
plt.bar(x1,y1)

x2=[5,7,6,9,7]
y2=[5,8,7,9,6]
plt.subplot(1,2,2)
plt.bar(x2,y2,color="red")
```

```python
x1=[5,8,7,6,9]
y1=[7,5,2,3,6]
plt.barh(x1,y1,color="green") #->horizontal representation
```

```python
#Histogram
#frequency of observation

x1=[1,1,5,5,6,3,8,5,2,2,2,4,8,8,8,9,5,4]
plt.hist(x1)
```

```python
#pie chart

y=[10,20,30,40]
mylabel=['apple','orange','mango','grapes']
plt.pie(y,labels=mylabel)
plt.legend(loc='lower right')  #anhor
```
