```python
import pandas as pd
df=pd.read_csv("D:\datascience\Iris.csv")
df
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
x_train.head()
```

```python
from sklearn import tree
#import decision tree classsifier
dt=tree.DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
score=accuracy_score(y_pred,y_test)
print(score)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
tree.plot_tree(dt,feature_names=['id','sepal_length','sepal_width','petal_length','petal_width'],class_names=['Iris-versicolor','Iris-virginica','Iris-setosa'],filled=True)
```
