```python
import numpy as np
import matplotlib.pyplot as plt
x=[1,5,7,8,9,5]
plt.plot(x)
plt.show
```

```python
y=np.array([1,5,2,7])
plt.plot(y,"*--y")  #*->marker  :->line  g:colour
plt.show()
```

```python
#line style property
plt.plot(y,linestyle='dotted')
```

```python
plt.plot(y,linewidth='3')
```

```python
plt.plot(y,linestyle='dotted',linewidth='7')
```

```python
x1=[1,2,3,4]
y1=[3,7,1,9]
x2=[3,4,6,1]
y2=[1,7,2,8]
plt.plot(x1,y1)
plt.plot(x2,y2)
```

```python
x1=[1,5,7,8,9,10]
y1=[7,8,9,4,5,2]
plt.plot(x1,y1)
plt.xlabel("Height in cm",color="g")   #fontsize=5
plt.ylabel("Weight in kg",color="r")   #fontsize=5
plt.title("Height-Weight graph")     #loc="left"-hedding position
plt.grid()   #grid(axis='x')
             #grid(axis='y')
```

```python
x2=[2,5,6,4,8,5]
y2=[6,4,5,8,9,2]
plt.plot(x2,y2)
plt.xlabel("velocity")
plt.ylabel("accelaration")
```

```python
x1=[1,5,7,8,9,10]
y1=[7,8,9,4,5,2]
plt.plot(x1,y1)
plt.xlabel("time")
plt.ylabel("displacement")
plt.title("velocity")
plt.grid(color='green',linestyle='--',linewidth='1')
```
